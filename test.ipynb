```python
import tensorflow as tf
from pathlib import Path

data_dir = r"s:\Projects\deep learning project\archive (1)\animals\animals"

dataset = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    labels="inferred",
    label_mode="int",       # "int", "categorical", or "binary"
    batch_size=32,
    image_size=(224, 224),  # resizing
    shuffle=True,
    seed=123
)

class_names = dataset.class_names
print("Classes:", class_names)  # mapping index -> class name
# example: get one batch
images, labels = next(iter(dataset))
print(images.shape, labels.shape)
```
